```python
import plotly.graph_objects as go
clef = 'pk.eyJ1IjoiZGpvdWwxMDExIiwiYSI6ImNsZTJ6MjdnODAzczczcXFqMHRwaTd6aHUifQ.S5xMsn0coaWPQFzfhPR-Cg'
```

```python
disques = go.Scattermapbox(
    lon=T['lon'],
    lat=T['lat'],
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=3 * T['morts'],
        color=['red'] * len(T),
    )
)

fig=go.Figure(disques)

# Specify the parameters for update_layout:
fig.update_layout(
    mapbox={
        'accesstoken': clef,
        'style': "streets",
        'center': {'lat': T['lat'].mean(), 'lon': T['lon'].mean()},
        'zoom': 15
    },
    margin={'r': 0, 'l': 0, 'b': 0}
)

fig.show()
```
